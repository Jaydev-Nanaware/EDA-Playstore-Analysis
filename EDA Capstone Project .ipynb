## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
psd_df.drop_duplicates(inplace=True)   # we are avoiding identicle rows for better results
````

````python
psd_df
````

````python
review_df.drop_duplicates(inplace=True)
review_df                         # we are avoiding identicle rows for better results
````

````python
# The + and , need to be replaced in order to typecast Installs column
psd_df['Installs'] = psd_df['Installs'].map(lambda x: x.rstrip('+')) 
psd_df['Installs'] = pd.to_numeric(psd_df['Installs'].str.replace(',',''))
````

````python
# It had some values containing $ so we replaced it and typecasted the Reviews 
psd_df['Reviews'] = pd.to_numeric(psd_df['Reviews'].str.replace('$',''))
````

````python
psd_df=psd_df.loc[psd_df.groupby(['App'])['Reviews'].idxmax()]
psd_df
````

````python
# we are creating top genres on the basis of number of apps in that genre
top_genres = psd_df.Genres.value_counts().reset_index().rename(columns={'Genres':'Count','index':'Genres'})
top_genres
````

````python
genres_installs = psd_df.groupby(['Genres'])[['Installs']].sum()
genres_installs
````

````python
psd_df['Price']=psd_df['Price'].str.replace("$","") 
psd_df['Price']=psd_df['Price'].astype(float)
````

````python
psd_df[psd_df['Price']>=50]
````

````python
# The difference between total installs is vaste so we have used log for proper comparison
psd_df['log_installs'] = np.log10(psd_df['Installs'])
````

````python
top_genres_installs = pd.merge(top_genres, genres_installs, on='Genres')
top_20_genres_installs = top_genres_installs.head(20)
````

````python
top_category = psd_df.Category.value_counts().reset_index().rename(columns={'Category':'Count','index':'Category'})
````

````python
category_installs = psd_df.groupby(['Category'])[['Installs']].sum()
````

````python
top_category_installs = pd.merge(top_category, category_installs, on='Category')
top_20_category_installs = top_category_installs
````

````python
genres_ratings_df = psd_df.groupby(['Genres'])[['Rating']].mean()
genres_installs_ratings = pd.merge(top_genres_installs, genres_ratings_df, on='Genres')
genres_installs_ratings['Rating'].describe()
````

````python
genres_installs_ratings.sort_values('Rating', ascending =False, inplace=True)
highest_rated_genres = genres_installs_ratings.iloc[0:20]
lowest_rated_genres = genres_installs_ratings.iloc[-20:]
lowest_rated_genres = lowest_rated_genres[lowest_rated_genres['Rating'].notnull()]
````

````python
app_count = psd_df.groupby(['Category','Type'])[['App']].count().reset_index().rename(columns={'App':'Count','index':'App'})
df_app_count = app_count.pivot('Category', 'Type', 'Count').fillna(0).reset_index()
````

````python
psd_df['Gaming Category App'] = psd_df['Category']=='GAME'
categoty_type_installs = psd_df.groupby(['Category','Type'])[['Installs']].sum().reset_index()
categoty_type_installs['log_Installs'] = np.log10(categoty_type_installs['Installs'])
````

````python
psd_df['Size'] = psd_df['Size'].apply(lambda x: str(x).replace('Varies with device', 'NaN') if 'Varies with device' in str(x) else x)
psd_df['Size'] = psd_df['Size'].apply(lambda x: str(x).replace('M', '') if 'M' in str(x) else x)
psd_df['Size'] = psd_df['Size'].apply(lambda x: str(x).replace(',', '') if 'M' in str(x) else x)
psd_df['Size'] = psd_df['Size'].apply(lambda x: float(str(x).replace('k', '')) / 1000 if 'k' in str(x) else x)
psd_df['Size'] = psd_df['Size'].apply(lambda x: float(x))
````

````python
psd_df.loc[psd_df['log_installs']==psd_df['log_installs'].min(),'log_installs']=0
````

````python
psd_df.loc[psd_df['Size'].isnull(),'Size']=0
````

````python
merged_df = psd_df.merge(review_df, on="App")
````

````python
category_sentiment = merged_df.groupby(['Category','Sentiment']).size().reset_index(name='Sentiment Count')
````

````python
category_sentiment['log_sentiment_count'] = np.log2(category_sentiment['Sentiment Count'])
````

````python
sentimet_subjectivity_polarity = merged_df.groupby(['Category'])['Sentiment_Polarity','Sentiment_Subjectivity'].mean().reset_index().rename(columns={'Sentiment_Polarity':'Sentiment_polarity_avg','Sentiment_Subjectivity':'sentiment_subjectivity_avg'})
````

````python
merged_df['Sentiment_Subjectivity'] = merged_df['Sentiment_Subjectivity'].abs()
merged_df['Sentiment_Polarity'] = merged_df['Sentiment_Polarity'].abs()
````

````python
review_notnull = merged_df[merged_df['Translated_Review'].notnull()]
````
